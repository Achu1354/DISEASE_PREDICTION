```python
nunique_cols = df.nunique() > 10
nunique_cols
```

```python
cols_to_transform = df.loc[:, nunique_cols].columns
cols_to_transform
```

```python
y.value_counts(normalize=True)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
x_smote, y_smote = smote.fit_resample(X , y)
```

```python
x_smote
```

```python
counts = y_smote.value_counts(normalize=True)
plt.pie(counts, labels = ["Risk", "No Risk"], autopct = "%1.1f%%")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size = 0.2)
```

```python
print(f"""
rows in X_train: {X_train.shape[0]}
rows in y_train: {y_train.shape[0]}

rows in X_test : {X_test.shape[0]}
rows in y_test : {y_test.shape[0]}

""")
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

scaler.fit(X_train[cols_to_transform])

X_train[cols_to_transform] = scaler.transform(X_train[cols_to_transform])
X_test[cols_to_transform] = scaler.transform(X_test[cols_to_transform])
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
GN = GaussianNB()

GN.fit(X_train, y_train)
GN_score = GN.score(X_test, y_test)
print(GN_score * 100)
```

```python
LR = LogisticRegression(max_iter=10000)

LR.fit(X_train, y_train)
LR_score = LR.score(X_test, y_test)
print(LR_score * 100)
```

```python
DTC = DecisionTreeClassifier()

DTC.fit(X_train, y_train)
DTC_score = DTC.score(X_test, y_test)
print(DTC_score * 100)
```

```python
RFC = RandomForestClassifier()

RFC.fit(X_train, y_train)
RFC_score = RFC.score(X_test, y_test)
print(RFC_score * 100)
```

```python
KNC = KNeighborsClassifier()

KNC.fit(X_train, y_train)
KNC_score = KNC.score(X_test, y_test)
print(KNC_score * 100)
```

```python
df_results = pd.DataFrame({
    "GN" : [GN_score],
    "LR" : [LR_score],
    "DTC": [DTC_score],
    "RFC": [RFC_score],
    "KNC" : [KNC_score]
})

df_results
```

```python
GN_predictions = GN.predict(X_test)
LR_predictions = LR.predict(X_test)
DTC_predictions = DTC.predict(X_test)
RFC_predictions = RFC.predict(X_test)
KNC_predictions = KNC.predict(X_test)

model_predictions_df = pd.DataFrame({
    "Guassian_NB" : GN_predictions,
    "LogisticRegression" : LR_predictions,
    "DecisionTreeClassifier" : DTC_predictions,
    "RandomForestClassifier" : RFC_predictions,
    "KNeighborsClassifier" : KNC_predictions
})

model_predictions_df
```

```python
cols = model_predictions_df.columns
prediction_counts = {col: model_predictions_df[col].value_counts() for col in cols}

actual_counts =y_test.value_counts()

print("Value Counts of Actual Labels (y_test):")
print(actual_counts)
print()

for col, counts in prediction_counts.items():
    print(f"Value Counts of Predictions for {col}:")
    print(counts)
    print()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(actual_counts, labels=actual_counts.index, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Pie Chart of Actual Labels (y_test)')
plt.show()

for col, counts in prediction_counts.items():
    plt.figure(figsize=(10, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
    plt.title(f'Pie Chart of Predictions for {col}')
    plt.show()
```

```python
from sklearn.metrics import accuracy_score, \
                            f1_score, precision_score, recall_score


model_evaluation = {}


for model, preds in model_predictions_df.items():
  model_evaluation[model] = [
        round(accuracy_score(y_test, preds) * 100, 2) ,
        round(f1_score(y_test, preds) * 100, 2),
        round(precision_score(y_test, preds) * 100, 2) ,
        round(recall_score(y_test, preds) * 100, 2),
    ]
```

```python
results_df = pd.DataFrame(model_evaluation, index=["Accuracy", "F1 Score", "Precision", "Recall"])

results_df.style.background_gradient(axis=None, cmap="bone")
```

```python
styled_df = results_df.style.background_gradient(axis=None, cmap="bone")

styled_df = styled_df.data

plt.figure(figsize=(10, 6))
plt.imshow(styled_df, aspect='auto', cmap="bone")
plt.xticks(range(len(results_df.columns)), results_df.columns, rotation=45, ha='right')
plt.yticks(range(len(results_df.index)), results_df.index)
plt.colorbar(label='Value')
plt.title('Model Comparison')
plt.tight_layout()
plt.show()
```
